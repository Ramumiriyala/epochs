```python
import numpy as np
# for AND gate (Input/Output)
x_list = np.array([[0,0],[0,1],[1,0],[1,1]])
y_list = np.array([0,0,0,1])
epochs = 6
w,b = np.zeros(2),0

for _ in range(epochs):
  for x,y in zip(x_list,y_list):
    z = np.dot(x,w) + b
    ### condition area for y_pred
    y_pred = 1 if z >= 0 else 0
    E = y - y_pred
    print("error : ",E,"<- at epoch",_)
    w += E*x
    b += E
    print(w,b)
```

```python
import numpy as np

X = np.array([[0,0],[0,1],[1,0],[1,1]])
Y = np.array([0,0,0,1])   # change for OR/XOR

w = np.random.randn(2)
b = np.random.randn()

for _ in range(20):
    err = 0
    for x,y in zip(X,Y):
        p = 1 if np.dot(x,w)+b >= 0 else 0
        e = y-p
        w += e*x; b += e
        err += abs(e)
    if err==0: break

print(w,b)
```
